```python
from tictactoe import Q_System_QL, plot_cnt_trace_normal_order_detail
import numpy as np

def learning_stage_qlearn(N_episodes=100, epsilon=0.4, fig_flag=False):
    
    N_Symbols = 3 # 0=empty, 1=plyaer1, 2=player2
    N_A = 9 # (0,0), (0,1), ..., (2,2)
    print_cnt = N_episodes / 5

    my_Q_System = Q_System_QL(N_A, N_Symbols)
    cnt_trace = my_Q_System.learning_qlearn(N_episodes=N_episodes, 
       epsilon=epsilon, print_cnt=print_cnt)
    my_Q_System.save()

    if fig_flag:
        plot_cnt_trace_normal_order_detail(cnt_trace, title='Q-learning')

    return my_Q_System
```
