```python
from tictactoe import learning_stage_qlearn

def q1_learning():
    # 강화학습의 전체 훈련 횟수 정의
    N_episodes = 1000
    # 탐험(Exploration) 빈도 정의
    epsilon = 0.2
    learning_stage_qlearn(N_episodes=N_episodes, 
        epsilon=epsilon, fig_flag=True, fig_mode='Basic')
```

```python
from tictactoe import Q_System_QL, plot_cnt_trace_normal_order_detail
import numpy as np

def learning_stage_qlearn(N_episodes=100, epsilon=0.4, save_flag=True,
                          fig_flag=False, fig_mode='Full'):
    ff = 0.9
    lr = 0.01
    N_Symbols = 3 # 0=empty, 1=plyaer1, 2=player2
    N_A = 9 # (0,0), (0,1), ..., (2,2)
    print_cnt = N_episodes / 5

    my_Q_System = Q_System_QL(N_A, N_Symbols)
    cnt_trace = my_Q_System.learning_qlearn(N_episodes=N_episodes, 
       ff=ff, lr=lr, epsilon=epsilon, print_cnt=print_cnt)

    if save_flag:
        my_Q_System.save()

    if fig_flag:
        plot_cnt_trace_normal_order_detail(cnt_trace, 
           title='Q-learning', fig_mode=fig_mode)

    return my_Q_System
```

```python
main()
```
